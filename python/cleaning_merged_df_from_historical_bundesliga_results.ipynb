```python
# Check unique team names in both datasets again
buli_home_teams = set(data['home_team_name'].unique())
buli_away_teams = set(data['away_team_name'].unique())
merged_home_teams = set(merged_df['home_team_name'].unique())
merged_away_teams = set(merged_df['away_team_name'].unique())
```

```python
# Find differences in home team names between the datasets
home_teams_buli_not_merged = buli_home_teams - merged_home_teams
home_teams_merged_not_buli = merged_home_teams - buli_home_teams
```

```python
# Find differences in away team names between the datasets
away_teams_buli_not_merged = buli_away_teams - merged_away_teams
away_teams_merged_not_buli = merged_away_teams - buli_away_teams
```

```python
# Define a mapping of current names to new names
name_mapping = {
    '1899 Hoffenheim': 'TSG 1899 Hoffenheim',
    'TSG Hoffenheim': 'TSG 1899 Hoffenheim',
    'Bayer 04 Leverkusen': 'Bayer Leverkusen',
    'Bayern München': 'FC Bayern München',
    'DSC Arminia Bielefeld': 'Arminia Bielefeld',
    'FC Augsburg 1907': 'FC Augsburg',
    'FSV Mainz 05': '1. FSV Mainz 05',
    'Hertha BSC Berlin': 'Hertha BSC',
    'SV Werder Bremen': 'Werder Bremen',
    'Schalke 04': 'FC Schalke 04',
    'Sport-Club Freiburg': 'SC Freiburg',
    'VfL Bochum 1848': 'VfL Bochum'
}

# Replace the home team names
merged_df['home_team_name'] = merged_df['home_team_name'].replace(name_mapping)

# Do the same for the away team names
merged_df['away_team_name'] = merged_df['away_team_name'].replace(name_mapping)
```
